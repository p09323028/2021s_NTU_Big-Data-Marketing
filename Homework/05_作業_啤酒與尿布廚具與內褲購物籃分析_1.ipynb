# **Create NextBuy Table**

```
nbt = creditlog.copy().sort_values(['客戶ID','刷卡日期'])
nbt['shift'] = (nbt['客戶ID'] == nbt['客戶ID'].shift(-1))
nbt['下次刷卡產品產業分類'] = nbt['刷卡產品產業分類'].shift(-1)
nbt = nbt[nbt['shift'] == True]
nbt
```

## **Probabilities of Next Purchase**

```
pnp = nbt.groupby(['刷卡產品產業分類','下次刷卡產品產業分類']).count()['刷卡類型'].reset_index()
pnp.set_index('刷卡產品產業分類', inplace=True)
pnp.rename(columns = {'刷卡類型' : '次數'}, inplace = True)
```

```
sumall = pnp['次數'].sum()
pnp['機率'] = pnp['次數'] / sumall
```
